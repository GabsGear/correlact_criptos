```python
from scipy.stats import linregress

x = np.arange(0, 500)
col = df.columns[1:len( df.columns)-1]

for column in col:
    value = linregress(x, df[column])
    print('MOEDA: ' + str(column))
    print(str(value) + '\n')
```

# Parametros da reta

slope : float
slope of the regression line

intercept : float
intercept of the regression line

rvalue : float
correlation coefficient

pvalue : float
two-sided p-value for a hypothesis test whose null hypothesis is that the slope is zero, using Wald Test with t-distribution of the test statistic.

stderr : float
Standard error of the estimated gradient.

```python
fig = plt.figure(figsize=(30, 30))

ax6 = fig.add_subplot(334)
ax6.set_xlabel('Amostra')
ax6.set_ylabel('valor')
ax6.set_title('NEOBTC')
plt.plot(df['NEOBTC'])

ax7 = fig.add_subplot(335)
ax7.set_xlabel('Amostra')
ax7.set_ylabel('valor')
ax7.set_title('ETHBTC')
plt.plot(df['ETHBTC'])

plt.show()
```

```python
def normalize(timeseries):
    normalized = (timeseries-timeseries.min())/(timeseries.max()-timeseries.min())
    return normalized 
    
neo = normalize(df['NEOBTC'])
eth = normalize(df['ETHBTC'])
```

```python
plt.plot(neo, label='NEOBTC')
plt.plot(eth, label='ETHBTC')
plt.xlabel('Amostra')
plt.ylabel('Valor')
plt.title("Compare")
plt.legend()
plt.show()
```
